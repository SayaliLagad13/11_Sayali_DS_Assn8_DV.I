```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.shape
```

```python
df.describe()
```

```python
df.describe(include = 'object')
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(np.mean(df['Age']))
df
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

```python
sns.catplot(x ='Sex', hue='Survived', kind='count', data=df)
plt.show()
```

```python
sns.catplot(x ='Embarked', hue='Survived', kind='count', col='Pclass', data=df)
plt.show()
```

```python
group = df.groupby(['Pclass','Survived'])
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived, annot=True, fmt="d")
plt.show()
```

```python
sns.violinplot(x='Sex', y='Age', data=df)
plt.show()
```

```python
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df,)
plt.show()
```

```python
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df, split=True)
plt.show()
```

```python
sns.barplot(x='Sex', y='Age', data=df)
plt.show()
```

```python
sns.barplot(x='Sex', y='Age', hue='Survived', data=df)
plt.show()
```

```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=df,)
plt.show()
```

```python
df['Family_Size'] = 0
df['Family_Size'] = df['Parch'] + df['SibSp']

df['alone'] = 0
df.loc[df.Family_Size == 0, ' alone'] = 1

sns.factorplot(x='Family_Size', y='Survived', data=df)
plt.show()
```

```python
df['Family_Size'] = 0
df['Family_Size'] = df['Parch']+df['SibSp']
 
df['alone'] = 0
df.loc[df.Family_Size == 0, 'alone'] = 1
 
sns.factorplot(x ='Family_Size', y ='Survived', data = df)
plt.show()
```

```python
df['Fare_Range'] = pd.qcut(df['Fare'],4)

sns.barplot(x='Fare_Range', y='Survived', data=df)
plt.show()
```

```python
sns.stripplot(x ='Sex', y ='Age', data = df)
plt.show()
```
